```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st

#Диаграмма рассеивания

# Номер зачётки - начальное состояние генератора
np.random.seed(19130345)

# Размер (объём) выборки
N = 100

# Генератор случайных чисел с равномерным распределением
x = np.random.uniform(low=150, high=200,size=N)

#описание генератора равномерного распределения
np.random.uniform?

# Генератор случайных чисел с нормальным распределением 
# начальное состояние генератора - номер зачетки в обратном порядке
np.random.seed(54303191)
e = np.random.normal(loc=0, scale=10,size=N)

#описание генератора нормального распределения
np.random.normal?

#разброс вокруг прямой линии
y=-100+x+e

plt.figure(figsize=(12,6))
plt.title ("Диаграмма рассеивания")
plt.xlabel("Ось x")
plt.ylabel("Ось y")

#строим график
plt.scatter(x,y)

# Не выводить использованные данные на экран
plt.show()

#Корреляционный анализ

#Коэффициент линейной корреляции Пирсона
r = st.pearsonr(x,y)[0]

#Коэффициент корреляции рангов Спирмена
rho = st.spearmanr(x,y)[0]

#Коэффициент корреляции рангов Кендалла
tau = st.kendalltau(x,y)[0]

#Выводим результаты на экран по заданию
print(f"Коэффициент Пирсона: {r:.3f}")
print(f"Коэффициент Спирмена: {rho:.3f}")
print(f"Коэффициент Кендалла: {tau:.3f}")

#Регрессионный анализ

#Коэффициенты уравнения регрессии y(x)
results_yx = st.linregress(x,y)
a = results_yx.intercept
b = results_yx.slope


#Вывод уравнения регрессии по заданию
print('Уравнение регрессии:')
print('')
print('100*a +Σ'f"{x}"'=Σ'f"{y}")
print('Σ'f"{x}"'*a+Σ'f"{x*x}"'*b=Σ'f"{x*y}")
print('')
print('a='f"{a:.3f}")
print('b='f"{b:.3f}")

plt.figure(figsize=(12,6))
plt.title ("Диаграмма рассеивания и линия регрессии")

#Данные для линии регрессии
x_yx = np.array([150,200])
y_yx = a + b * x_yx
plt.plot(x_yx,y_yx, color='red', label='Линия регрессии')
plt.scatter(x,y,color='blue',label='Диаграмма рассеивания')
plt.legend()
plt.show()

# Коэффициенты уравнения регрессии x(y)
results_xy = st.linregress(y,x)
c = results_xy.intercept
d = results_xy.slope
c,d

#Данные для линии регрессии x(y)
x_xy = np.array([150,200])
y_xy = c + d * x_xy

#Средние значения
x_mean = np.mean(x)
y_mean = np.mean(y)
x_mean, y_mean


#Диаграмма рассеивания+линии регрессии+ пересечение
plt.figure(figsize=(12,6))
plt.title ("Диаграмма рассеивани, и линия регрессии, и пересечение")

plt.plot(x_yx,y_yx, color='green', label='Линия регрессии y(x)')
plt.plot(x_xy,y_xy, color='blue', label='Линия регрессии x(y)')
plt.scatter(x,y,color='grey',label='Исходные данные')
plt.scatter(x_mean,y_mean,color='red',label='Среднее значение')
plt.legend()
plt.show()
```
